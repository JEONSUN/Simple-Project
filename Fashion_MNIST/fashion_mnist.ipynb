```python
from tensorflow import keras
from sklearn.model_selection import train_test_split
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
train_df = pd.read_csv('D:/Portfolio/Simple Project Datasets/fashion mnist/train.csv')
test_df = pd.read_csv('D:/Portfolio/Simple Project Datasets/fashion mnist/test.csv')
```

```python
train_df.head()
```

```python
train_df.info()
```

```python
test_df.copy()
```

```python
train = train_df.copy()
td = train.drop(['index','label'],axis = 1)
tt = train['label']
ts = test_df.copy()
ts = ts.drop(['index'],axis = 1)
```

```python
class_names = ['T_shirt/top', 'Trouser', 'Pullover', 'Dress', 'Coat', 
               'Sandal', 'Shirt', 'Sneaker', 'Bag', 'Ankle boot']
```

```python
td = np.array(td)
ts = np.array(ts)
```

```python
ttd = td.reshape(-1,28,28,1)
tss = ts.reshape(-1,28,28,1)
```

```python
j = 1
plt.figure(figsize = (20,20))
for i in range(0,40) :
    plt.subplot(8,5,j);j +=1
    plt.imshow(ttd[i])
    plt.xticks([])
    plt.yticks([])
    plt.title('{} : {}'.format(tt[i],class_names[tt[i]]))
```

```python
td_scaled = ttd/255.0
ts_scaled = tss/255.0
td_scaled
```

```python
train_scaled,val_scaled, train_target,val_target = train_test_split(td_scaled, tt, test_size = 0.2, random_state = 42)
```

```python
train_scaled.shape
```

## 1. ANN 모델 구현

```python
def model1(a_layer = None) :
    model = keras.Sequential()
    model.add(keras.layers.Flatten(input_shape = (28,28,1)))
    model.add(keras.layers.Dense(100, activation = 'relu'))
    
    if a_layer :
        model.add(a_layer)
    model.add(keras.layers.Dense(10,activation = 'softmax'))
    return model
```

```python
model = model1(keras.layers.Dropout(0.3))
model.summary()
```

```python
model.compile(optimizer = 'adam',loss = 'sparse_categorical_crossentropy', metrics = 'accuracy')
history = model.fit(train_scaled,train_target, epochs = 7, verbose = 0, validation_data = (val_scaled,val_target))
fig, axs = plt.subplots(2,1, figsize = (10,8))
axs[0].plot(history.history['loss'])
axs[0].plot(history.history['val_loss'])
axs[1].plot(history.history['accuracy'])
axs[1].plot(history.history['val_accuracy'])
axs[0].legend(['train_loss','val_loss'])
axs[1].legend(['train_acc','val_acc'])
plt.show()
```

```python
pred = np.argmax(model.predict(ts_scaled),axis = -1)
```

## CNN(Convolutional Neural Network)모델

- 합성곱 신경망은 **합성곱(convolution)**과 **풀링(pooling)**을 통해서 이미지 데이터의 형상 정보를 유지하면서 학습하여 이미지 인식 및 처리의 성능을 향상시킨 신경망이다

- **합성곱층(convolutional layer) → 활성화 함수 적용 → 풀링층(pooling layer) 과정을 반복하며(Feature Learning)** 완전히 상호 연결된 결합된 은닉층이 추가되며, 최종적으로 출력층이 배치된다.

이때 **합성곱층**은 이미지 데이터의 특성을 추출하는 역할을 하며, **풀링층**은 추출된 특성 정보를 축소하는 역할을 한다.

```python
from keras.models import Sequential
from keras.layers import Dense,Flatten,Dropout
from keras.layers.convolutional import Conv2D, MaxPooling2D
```

### Convolution 층
- 이

### Maxpooling
- 구역의 대표값만 추출하여 출력의 공간적 크기를 줄인다.

### Model.add
- 2차원 데이터를 1차원으로 변환(입력층과 은닉층사이에 추가)

```python
# cnn모델 구현
model = Sequential()
model.add(Conv2D(filters = 32, kernel_size = (3,3),
                strides = (1,1), padding = 'Same',
                activation = 'relu', input_shape = (28,28,1)))
model.add(MaxPooling2D(pool_size = (2,2)))
model.add(Conv2D(filters = 64, kernel_size = (3,3),
                strides = (1,1), padding = 'Same',
                activation = 'relu'))
model.add(Conv2D(filters = 32, kernel_size = (3,3),
                strides = (1,1), padding = 'Same',
                activation = 'relu'))
model.add(MaxPooling2D(pool_size = (2,2)))
model.add(Flatten())
model.add(Dense(units = 1024, activation = 'relu'))
model.add(Dropout(0.5))
model.add(Dense(units = 1024, activation = 'relu'))
model.add(Dropout(0.5))
model.add(Dense(units = 10, activation = 'softmax'))
```

```python
# to_categorical로 변환 가능
model.compile(loss = 'sparse_categorical_crossentropy',
             optimizer = 'adam',
             metrics = ['accuracy'])
```

```python
history = model.fit(train_scaled,train_target,
         epochs = 10,
         batch_size = 128,
         validation_data = (val_scaled,val_target))
```

```python
val_labels = np.argmax(model.predict(val_scaled),axis = -1)
```

```python
fig,axs = plt.subplots(figsize = (12,6))
axs.set_xlabel('Epoch', fontsize = 14)
axs.set_ylabel('Loss',fontsize = 14)
axs.plot(history.history['loss'],label = 'train_loss',c = 'red')
axs.plot(history.history['val_loss'], label = 'valid_loss', c = 'green')
axs.legend(loc = 'upper right',fontsize = 14)
axs2 = axs.twinx()
axs2.set_ylabel('Accuracy',fontsize = 14)
axs2.plot(history.history['accuracy'], label = 'train_acc',linestyle = '--')
axs2.plot(history.history['val_accuracy'], label = 'valid_acc',linestyle = '--')
axs2.legend(loc = 'lower right',fontsize = 14)
plt.show()
```

```python
# heatmap으로 예측 결과 확인
from sklearn.metrics import confusion_matrix
import seaborn as sns

class_names = ['T_shirt/top', 'Trouser', 'Pullover', 'Dress', 'Coat', 
               'Sandal', 'Shirt', 'Sneaker', 'Bag', 'Ankle boot']

plt.figure(figsize = (16,9))

cm = confusion_matrix(val_target,val_labels)

sns.heatmap(cm , annot = True,fmt = 'd',xticklabels = class_names,yticklabels = class_names)
plt.show()
```
